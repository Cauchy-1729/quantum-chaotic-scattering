```python
# Cell 1: Imports and solver class
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.fft import fft, ifft, fftfreq
#from scipy.integrate import trapz
from IPython.display import HTML

class QuantumScatteringSolver:
    def __init__(self, L_total=120.0, N_points=1024, m=1.0, hbar=1.0):
        self.L = L_total
        self.N = N_points
        self.m = m
        self.hbar = hbar
        self.dx = L_total / (N_points - 1)
        self.x = np.linspace(-L_total/2, L_total/2, N_points)
        self.dp = 2 * np.pi / L_total
        self.p = hbar * fftfreq(N_points, d=self.dx) * 2 * np.pi
        self.dt = None
        self.t = 0.0
        self.psi = None
        self.V = None
        self.abc_width = int(N_points * 0.1)
        self.abc_strength = 0.01

    def create_blumel_potential(self, V0=2.0, sigma=3.0):
        positions = [-15.0, 0.0, 15.0]
        self.V = np.sum([V0 * np.exp(-(self.x - pos)**2 / (2 * sigma**2)) 
                        for pos in positions], axis=0)
        return self.V

    def create_absorbing_boundaries(self):
        abs_pot = np.zeros(self.N, dtype=complex)
        # Left boundary
        for i in range(self.abc_width):
            factor = (self.abc_width - i) / self.abc_width
            abs_pot[i] = -1j * self.abc_strength * factor**2
        # Right boundary
        for i in range(self.N - self.abc_width, self.N):
            factor = (i - (self.N - self.abc_width - 1)) / self.abc_width
            abs_pot[i] = -1j * self.abc_strength * factor**2
        return abs_pot

    def initialize_incoming_wave(self, k0=1.5, x0=-40.0, sigma_x=8.0):
        self.psi = np.exp(-(self.x - x0)**2 / (4 * sigma_x**2)) * np.exp(1j * k0 * self.x)
        norm = np.sqrt(np.trapz(np.abs(self.psi)**2, self.x))
        # self.psi /= norm
        # Set stable time step
        dt_stability = 0.4 * self.m * self.dx**2 / self.hbar
        if self.V is not None:
            dt_stability = min(dt_stability, 0.1 * self.hbar / np.max(self.V))
        self.dt = dt_stability
        return self.psi.copy()
        
        

    def split_operator_step(self):
        # Potential half-step
        if self.V is not None:
            V_total = self.V + self.create_absorbing_boundaries()
            self.psi *= np.exp(-1j * V_total * self.dt / (2 * self.hbar))
        
        # Kinetic full-step (momentum space)
        psi_k = fft(self.psi)
        kinetic_phase = np.exp(-1j * self.p**2 * self.dt / (2 * self.m * self.hbar))
        self.psi = ifft(psi_k * kinetic_phase)
        
        # Potential half-step
        if self.V is not None:
            V_total = self.V + self.create_absorbing_boundaries()
            self.psi *= np.exp(-1j * V_total * self.dt / (2 * self.hbar))
        
        self.t += self.dt

    def evolve_with_snapshots(self, total_time, snapshot_interval=0.5):
        n_steps = int(total_time / self.dt)
        steps_per_snapshot = max(1, int(snapshot_interval / self.dt))
        snapshots = []
        times = []
        
        for step in range(n_steps):
            self.split_operator_step()
            if step % steps_per_snapshot == 0:
                snapshots.append(self.psi.copy())
                times.append(self.t)
                
        return np.array(times), np.array(snapshots)

    def analyze_transmission(self, psi, psi_initial):
        """Calculate transmission coefficient from final wavefunction"""
        # Define asymptotic regions
        left_mask = self.x < -30.0
        right_mask = self.x > 30.0
        
        # Calculate probability currents
        def current(psi, x):
            dpsi = np.gradient(psi, x[1]-x[0])
            return (self.hbar / self.m) * np.imag(np.conj(psi) * dpsi)
        
        J_left = np.mean(np.abs(current(psi[left_mask], self.x[left_mask])))
        J_right = np.mean(np.abs(current(psi[right_mask], self.x[right_mask])))
        
        # Incident current (from initial wave)
        k0 = 1.5  # Should match initialization
        # J_incident = self.hbar * k0 / self.m
        J_incident = np.mean(np.abs(current(psi_initial[left_mask], self.x[left_mask])))
        
        print(J_incident)
        print(J_left)
        print(J_right)
        
        T = abs(J_right) / J_incident
        R = abs(J_left) / J_incident
        return T, R, T + R
```

```python
# Cell 2: Visualization functions
def plot_initial_state(solver):
    """Plot initial wavefunction and potential"""
    plt.figure(figsize=(10, 6))
    plt.plot(solver.x, np.real(solver.psi), 'b-', label='Real(ψ)')
    plt.plot(solver.x, np.imag(solver.psi), 'r-', label='Imag(ψ)')
    plt.plot(solver.x, np.abs(solver.psi), 'k-', label='|ψ|')
    plt.plot(solver.x, solver.V, 'm-', label='Potential')
    plt.xlabel('Position')
    plt.ylabel('Amplitude')
    plt.title('Initial Wavefunction and Potential')
    plt.legend()
    plt.grid(True)
    plt.show()

def create_animation(solver, times, snapshots):
    """Create and display wavefunction evolution animation"""
    plt.rcParams['animation.embed_limit'] = 70
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Initialize plot elements
    line_real, = ax.plot(solver.x, np.real(snapshots[0]), 'b-', label='Real(ψ)')
    line_imag, = ax.plot(solver.x, np.imag(snapshots[0]), 'r-', label='Imag(ψ)')
    line_abs, = ax.plot(solver.x, np.abs(snapshots[0]), 'k-', label='|ψ|')
    line_pot, = ax.plot(solver.x, solver.V, 'm-', label='Potential')
    
    ax.set_xlim(solver.x[0], solver.x[-1])
    ax.set_ylim(-0.5, 1.5)
    ax.set_xlabel('Position')
    ax.set_ylabel('Wavefunction')
    ax.set_title(f'Time = {times[0]:.2f}')
    ax.legend(loc='upper right')
    ax.grid(True)

    def update(frame):
        line_real.set_ydata(np.real(snapshots[frame]))
        line_imag.set_ydata(np.imag(snapshots[frame]))
        line_abs.set_ydata(np.abs(snapshots[frame]))
        ax.set_title(f'Time = {times[frame]:.2f}')
        return line_real, line_imag, line_abs

    ani = FuncAnimation(fig, update, frames=len(times), interval=50, blit=True)
    plt.close()
    return HTML(ani.to_jshtml())

def plot_final_analysis(solver, final_psi):
    """Plot final state with transmission analysis"""
    plt.figure(figsize=(12, 8))
    
    # Wavefunction and potential
    plt.subplot(211)
    plt.plot(solver.x, np.real(final_psi), 'g-', label='Re(ψ)')
    plt.plot(solver.x, solver.V, 'm-', label='Potential')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    
    # Phase and current analysis
    plt.subplot(212)
    plt.plot(solver.x, np.angle(final_psi), 'b-', label='Phase')
    
    # Highlight asymptotic regions
    plt.axvspan(solver.x[0], -30, alpha=0.1, color='red', label='Reflection Region')
    plt.axvspan(30, solver.x[-1], alpha=0.1, color='green', label='Transmission Region')
    
    plt.xlabel('Position')
    plt.ylabel('Phase (radians)')
    plt.legend()
    plt.grid(True)
    plt.suptitle('Final Wavefunction Analysis')
    plt.tight_layout()
    plt.show()

```

```python
# Cell 3: Main simulation
# Initialize solver
solver = QuantumScatteringSolver(L_total=1000.0, N_points=1024)

# Create Blümel potential
solver.create_blumel_potential(V0=1.2, sigma=3.0)

# Initialize wavefunction (wide Gaussian ~ plane wave)
k0 = 1.5
psi_initial = solver.initialize_incoming_wave(k0=k0, x0=-70.0, sigma_x=8.0)

# Plot initial state
plot_initial_state(solver)

# Run simulation with snapshots for animation
print("Running quantum evolution...")
times, snapshots = solver.evolve_with_snapshots(total_time=300.0, snapshot_interval=0.5)
final_psi = snapshots[-1]

# Create and display animation
print("Generating animation...")
# animation.embed_limit['rc'] = 25.0
# animation_html = create_animation(solver, times, snapshots)
# display(animation_html)

# Plot final analysis
plot_final_analysis(solver, final_psi)

# Calculate transmission coefficient
T, R, conservation = solver.analyze_transmission(final_psi, psi_initial)
print(f"\nSCATTERING RESULTS:")
print(f"Transmission coefficient (T): {T:.4f}")
print(f"Reflection coefficient (R): {R:.4f}")
print(f"Conservation check (T+R): {conservation:.4f}")

```
