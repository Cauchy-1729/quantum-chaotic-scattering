```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags, block_diag
from scipy.linalg import eigh

# Parameters (use same as your solver)
L_total = 120.0
N_points = 512
x = np.linspace(-L_total/2, L_total/2, N_points)
dx = x[1] - x[0]
hbar = 1.0
m = 1.0
I_0 = 5.0
l_channels = np.arange(-2, 3)
n_channels = len(l_channels)
V0 = 3.0
xi = 10.0

# 1. Construct the spatial potential
positions = [-xi, 0.0, xi]
V_spatial = np.sum([V0/2 * np.exp(-(x - pos)**2) for pos in positions], axis=0)

# 2. Build the Hamiltonian in the angular momentum basis
def build_hamiltonian(l_channels, x, V_spatial, hbar=1.0, m=1.0, I_0=5.0):
    N = len(x)
    n_channels = len(l_channels)
    # Kinetic energy (finite difference Laplacian)
    main_diag = -2.0 * np.ones(N) / dx**2
    off_diag = np.ones(N-1) / dx**2
    laplacian = diags([main_diag, off_diag, off_diag], [0, -1, 1]).toarray()
    T = -(hbar**2)/(2*m) * laplacian  # (N,N)
    # Potential (diagonal in x)
    Vmat = np.diag(V_spatial)
    # Rotational energy (diagonal in l)
    rot_energies = [(hbar**2 * l**2)/(2*I_0) for l in l_channels]
    # Hamiltonian blocks
    H_blocks = []
    for i, l in enumerate(l_channels):
        H_l = T + Vmat + np.eye(N) * rot_energies[i]
        H_blocks.append(H_l)
    # Inter-channel coupling (nearest neighbor in l)
    V_coupling = []
    for i in range(n_channels-1):
        V_coupling.append(np.diag(V_spatial))
    # Build full Hamiltonian (block tridiagonal)
    H = np.zeros((n_channels*N, n_channels*N), dtype=np.complex128)
    for i in range(n_channels):
        H[i*N:(i+1)*N, i*N:(i+1)*N] = H_blocks[i]
        # Coupling to l+1 and l-1
        if i < n_channels-1:
            H[i*N:(i+1)*N, (i+1)*N:(i+2)*N] = np.diag(V_spatial)
            H[(i+1)*N:(i+2)*N, i*N:(i+1)*N] = np.diag(V_spatial)
    return H

H = build_hamiltonian(l_channels, x, V_spatial, hbar, m, I_0)

# 3. Diagonalize (for large matrices, use sparse eigensolvers for a subset)
n_eigs = 50  # Number of eigenvalues/eigenfunctions to compute
eigvals, eigvecs = eigh(H)
eigvals = eigvals[:n_eigs]
eigvecs = eigvecs[:, :n_eigs]

N = len(x)

# 4. Plot a few eigenfunctions (as a function of x, for each channel)
plt.figure(figsize=(160, 80))
for idx in range(50):
    plt.subplot(2, 25, idx+1)
    for i, l in enumerate(l_channels):
        psi_l = eigvecs[i*N:(i+1)*N, idx]
        plt.plot(x, np.real(psi_l), label=f'l={l}')
    plt.title(f'Eigenfunction #{idx+1}, E={eigvals[idx]:.3f}')
    plt.xlabel('x')
    plt.ylabel('Re[Ïˆ(x)]')
    plt.legend()
plt.tight_layout()
plt.show()

# 5. Level spacing statistics
sorted_eigvals = np.sort(eigvals.real)
spacings = np.diff(sorted_eigvals)
mean_spacing = np.mean(spacings)
norm_spacings = spacings / mean_spacing

plt.figure(figsize=(7,5))
plt.hist(norm_spacings, bins=20, density=True, alpha=0.7, label='Numerical')
s = np.linspace(0, 4, 100)
poisson = np.exp(-s)
wigner = (np.pi/2) * s * np.exp(-np.pi * s**2 / 4)
plt.plot(s, poisson, 'r--', label='Poisson')
plt.plot(s, wigner, 'g-', label='Wigner-Dyson')
plt.xlabel('Normalized Level Spacing $s$')
plt.ylabel('P(s)')
plt.title('Level Spacing Distribution')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L_total = 40.0
N_points = 1000
x = np.linspace(-L_total/2, L_total/2, N_points)
dx = x[1] - x[0]
hbar = 1.0
m = 1.0
I_0 = 5.0
l_channels = np.arange(-2, 3)
n_channels = len(l_channels)
V0 = 3.0
xi = 10.0

# Potential (same for all channels, but coupling will be added)
positions = [-xi, 0.0, xi]
V_spatial = np.sum([V0/2 * np.exp(-(x - pos)**2) for pos in positions], axis=0)

def coupled_numerov(E, parity='even'):
    """
    Numerov integration for coupled channel system.
    Returns wavefunctions for all channels, integrated from left and right.
    """
    # Rotational energies for each channel
    E_rot = np.array([(hbar**2 * l**2) / (2*I_0) for l in l_channels])
    # Effective potential for each channel
    V_eff = np.array([V_spatial + E_rot[i] for i in range(n_channels)])
    # Coupling matrix: only nearest neighbor in l
    def coupling_mat(idx):
        mat = np.zeros((n_channels, n_channels))
        for i in range(n_channels):
            if i > 0:
                mat[i, i-1] = V_spatial[idx]
            if i < n_channels-1:
                mat[i, i+1] = V_spatial[idx]
        return mat

    # Numerov matrices
    k2 = 2*m*(E - V_eff) / hbar**2  # shape: (n_channels, N_points)
    psi_left = np.zeros((n_channels, N_points))
    psi_right = np.zeros((n_channels, N_points))
    
    # Initial conditions for left integration (x[0], x[1])
    if parity == 'even':
        psi_left[:, 0] = 1e-8  # all channels small, even parity
        psi_left[n_channels//2, 0] = 1.0  # central channel dominant
        psi_left[:, 1] = psi_left[:, 0]  # even: symmetric
    else:  # odd
        psi_left[:, 0] = 0.0
        psi_left[n_channels//2, 1] = 1e-5  # odd: antisymmetric

    # Numerov integration from left
    for i in range(1, N_points-1):
        # Coupling at i
        C = coupling_mat(i)
        # Standard Numerov for each channel, with coupling
        psi_left[:, i+1] = (2*psi_left[:, i] - psi_left[:, i-1] + 
            dx**2 * (k2[:, i]*psi_left[:, i] + np.dot(C, psi_left[:, i])))
    
    # Initial conditions for right integration (x[-1], x[-2])
    if parity == 'even':
        psi_right[:, -1] = 1e-8
        psi_right[n_channels//2, -1] = 1.0
        psi_right[:, -2] = psi_right[:, -1]
    else:
        psi_right[:, -1] = 0.0
        psi_right[n_channels//2, -2] = 1e-5

    # Numerov integration from right
    for i in range(N_points-2, 0, -1):
        C = coupling_mat(i)
        psi_right[:, i-1] = (2*psi_right[:, i] - psi_right[:, i+1] + 
            dx**2 * (k2[:, i]*psi_right[:, i] + np.dot(C, psi_right[:, i])))
    
    return psi_left, psi_right

def mismatch(E, parity='even'):
    """Compute mismatch at match point for all channels."""
    psi_left, psi_right = coupled_numerov(E, parity)
    match_idx = N_points // 2
    # Rescale right solution to match left at match point
    scale = psi_left[:, match_idx] / (psi_right[:, match_idx] + 1e-20)
    psi_right_scaled = psi_right * scale[:, None]
    # Compute derivative mismatch
    dpsi_left = (psi_left[:, match_idx+1] - psi_left[:, match_idx-1]) / (2*dx)
    dpsi_right = (psi_right_scaled[:, match_idx+1] - psi_right_scaled[:, match_idx-1]) / (2*dx)
    return np.sum(np.abs(dpsi_left - dpsi_right))

# Energy search
def find_eigenenergies(E_range, parity='even', tol=1e-5, max_iter=100):
    """Scan E_range for eigenenergies with minimal mismatch."""
    E_grid = np.linspace(*E_range, 200)
    mismatches = np.array([mismatch(E, parity) for E in E_grid])
    # Find local minima
    from scipy.signal import argrelextrema
    minima = argrelextrema(mismatches, np.less)[0]
    eigenEs = []
    for idx in minima:
        E0 = E_grid[idx]
        # Refine using a local minimizer
        from scipy.optimize import minimize_scalar
        res = minimize_scalar(mismatch, bracket=(E0-0.05, E0+0.05), args=(parity,), tol=tol)
        if res.success and res.fun < 1e-3:
            eigenEs.append(res.x)
    return eigenEs

# Example: search for even and odd eigenenergies in two regions
E_even_1 = find_eigenenergies((0, 1.5), parity='even')
E_odd_1 = find_eigenenergies((0, 1.5), parity='odd')
E_even_2 = find_eigenenergies((1.5, 2.3), parity='even')
E_odd_2 = find_eigenenergies((1.5, 2.3), parity='odd')

print("Even eigenenergies (0,1.5):", E_even_1)
print("Odd eigenenergies (0,1.5):", E_odd_1)
print("Even eigenenergies (1.5,2.3):", E_even_2)
print("Odd eigenenergies (1.5,2.3):", E_odd_2)

# Plot an eigenfunction (example)
E_plot = E_even_1[0]
psi_left, psi_right = coupled_numerov(E_plot, parity='even')
match_idx = N_points // 2
scale = psi_left[:, match_idx] / (psi_right[:, match_idx] + 1e-20)
psi_combined = np.zeros_like(psi_left)
for ch in range(n_channels):
    psi_combined[ch, :match_idx+1] = psi_left[ch, :match_idx+1]
    psi_combined[ch, match_idx+1:] = psi_right[ch, match_idx+1:] * scale[ch]

plt.figure(figsize=(10,6))
for ch in range(n_channels):
    plt.plot(x, psi_combined[ch], label=f'l={l_channels[ch]}')
plt.title(f'Coupled eigenfunction at E={E_plot:.4f}')
plt.xlabel('x')
plt.legend()
plt.show()
```
